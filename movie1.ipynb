```python
# Create en new moviesdb.db

import sqlite3
 
# create a new db
db = sqlite3.connect("BDD/moviesdb.db")   
cursor = db.cursor()    

cursor.execute('''create table movies ( 
movieId integer primary key,
title text,
genre text)'''
)

cursor.execute('''create table links ( 
movieId integer,
imdbId integer primary key,
tmdbId integer)'''
) 

cursor.execute('''create table ratings ( 
userId integer,
movieId integer,
rating integer,
timestamp integer)'''
) 

cursor.execute('''create table tags ( 
userId integer,
movieId integer,
tag text,
timestamp integer)'''
) 

db.commit()   

cursor.close()

db.close() 
```

```python

# Insert data in tables's moviescd from csv files

import csv
import sqlite3

filename1 = 'movies csv/movies.csv'
filename2 = 'movies csv/links.csv'
filename3 = 'movies csv/ratings.csv'
filename4 = 'movies csv/tags.csv' 

# create a new db
db = sqlite3.connect("BDD/moviesdb.db")   
cursor = db.cursor()    


with open(filename1,'r', encoding='utf-8') as csv_file1:
    reader = csv.reader(csv_file1, delimiter=',')
    next(reader)
    
    for row1 in reader:
        cursor.execute('insert into movies (movieId, title, genre) values (?, ?, ?)', row1)
        
        
with open(filename2,'r', encoding='utf-8') as csv_file2:
    reader = csv.reader(csv_file2, delimiter=',')
    next(reader)
    
    for row2 in reader:
        cursor.execute('insert into links (movieId, imdbId, tmdbId) values (?, ?, ?)', row2)
        
with open(filename3,'r', encoding='utf-8') as csv_file3:
    reader = csv.reader(csv_file3, delimiter=',')
    next(reader)
    
    for row3 in reader:
        cursor.execute('insert into ratings (userId, movieId, rating, timestamp) values (?, ?, ?, ?)', row3)
        
with open(filename4,'r', encoding='utf-8') as csv_file4:
    reader = csv.reader(csv_file4, delimiter=',')
    next(reader)

    for row4 in reader:
        cursor.execute('insert into tags (userId, movieId, tag, timestamp) values (?, ?, ?, ?)', row4)
          

db.commit()   

cursor.close()

db.close()      
```
